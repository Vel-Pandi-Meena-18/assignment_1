```python
import requests
import pandas as pd
import time

# This list will hold our data as we collect it
all_records = []

# We want 5 pages of data (1, 2, 3, 4, 5)
for page in range(1, 6):

    # This is the address we are asking for data from
    url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&order=market_cap_desc&per_page=250&page={page}&sparkline=false"

    # We try to get the data
    response = requests.get(url)

    # Check if the connection was successful (200 means OK)
    if response.status_code == 200:
        data = response.json()
        all_records.extend(data)
        print(f"Success! Page {page} is finished.")

        # We wait 15 seconds so the website doesn't get angry at us
        time.sleep(15)

    # If the website says "Too Many Requests" (429), we wait longer
    elif response.status_code == 429:
        print("The website is busy. Waiting 40 seconds to try again...")
        time.sleep(40)
        # We try the same page again
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            all_records.extend(data)
            print(f"Got it! Page {page} is now finished.")
            time.sleep(15)
    else:
        print(f"Something went wrong on Page {page}. Error code: {response.status_code}")

# Now we turn that list into a neat table (DataFrame)
df_crypto = pd.DataFrame(all_records)

# Your project only needs these specific columns [cite: 83]
required_columns = [
    'id', 'symbol', 'name', 'current_price', 'market_cap',
    'market_cap_rank', 'total_volume', 'circulating_supply',
    'total_supply', 'ath', 'atl', 'last_updated'
]

# We filter the table to keep only what is required
df_crypto_final = df_crypto[required_columns].copy()

# We make sure the date looks like a normal date (YYYY-MM-DD) [cite: 83]
df_crypto_final['last_updated'] = pd.to_datetime(df_crypto_final['last_updated']).dt.date

print("---------------------------------------")
print(f"Done! You have collected {len(df_crypto_final)} rows of data.")
```

## Top 3 Historical Coins

```python
import requests
import pandas as pd
import time
from google.colab import files

# 1. Define the parameters for your Top 3 coins
# Based on market_cap_rank, these are the current leaders
top_3_ids = ['bitcoin', 'ethereum', 'tether']
days_count = '365'
currency = 'inr'

all_history = []

print(f"Starting extraction for {top_3_ids} in {currency.upper()}...")

for coin in top_3_ids:
    # The URL requested in your Step 2
    url = f"https://api.coingecko.com/api/v3/coins/{coin}/market_chart?vs_currency={currency}&days={days_count}&interval=daily"

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        # 'prices' comes as a list of [timestamp, price_value]
        prices = data['prices']

        for entry in prices:
            all_history.append({
                'coin_id': coin,
                'date': pd.to_datetime(entry[0], unit='ms').date(),
                'price_inr': entry[1]
            })
        print(f"Successfully collected 1 year of data for {coin}.")
        # Waiting to avoid being blocked by the API
        time.sleep(15)
    else:
        print(f"Error fetching {coin}. Status: {response.status_code}")

# 2. Create the DataFrame and Download
df_history = pd.DataFrame(all_history)
file_name = 'top_3_historical_prices_inr.csv'
df_history.to_csv(file_name, index=False)
files.download(file_name)

print("---------------------------------------")
print(f"DONE! {len(df_history)} total price points are now in your file.")
```

## Oil stocks

```python
import pandas as pd
from google.colab import files

# 1. Load the data from the exact URL provided
oil_url = "https://raw.githubusercontent.com/datasets/oil-prices/main/data/wti-daily.csv"
df_oil = pd.read_csv(oil_url)

# 2. Fix the dates and filter (Jan 2020 - Jan 2026)
# Note: The file uses 'Date' and 'Price' as column names
df_oil['date'] = pd.to_datetime(df_oil['Date']).dt.date
start_date = pd.to_datetime('2020-01-01').date()
end_date = pd.to_datetime('2026-01-31').date()

df_oil_filtered = df_oil[(df_oil['date'] >= start_date) & (df_oil['date'] <= end_date)].copy()

# 3. Convert to INR (using current approximate rate of 83.0)
usd_to_inr = 83.0
df_oil_filtered['oil_price_inr'] = df_oil_filtered['Price'] * usd_to_inr

# 4. Keep only the final columns for your project
df_oil_final = df_oil_filtered[['date', 'oil_price_inr']]

# 5. Save and Download
file_name = 'oil_prices_2020_2026_inr.csv'
df_oil_final.to_csv(file_name, index=False)
files.download(file_name)

print("---------------------------------------")
print(f"Success! {len(df_oil_final)} daily oil prices collected in INR.")
```

## Stock Prices (Yahoo Finance API)

```python
import yfinance as yf
import pandas as pd
from google.colab import files

# 1. Define the tickers and the date range
tickers = ["^GSPC", "^IXIC", "^NSEI"]  # S&P 500, NASDAQ, NIFTY 50
start_date = "2020-01-01"
end_date = "2025-09-30"

print(f"Fetching Stock data for {tickers}...")

# 2. Download daily close prices
# auto_adjust=True handles stock splits automatically
stock_data = yf.download(tickers, start=start_date, end=end_date)['Close']

# 3. Handle the "Multi-Index" and fix the dates
df_stocks = stock_data.reset_index()
df_stocks['Date'] = pd.to_datetime(df_stocks['Date']).dt.date
df_stocks = df_stocks.rename(columns={'Date': 'date'})

# 4. Convert USD Indices to INR (Rate: 83.0)
# NIFTY 50 (^NSEI) is already in INR, so we don't touch it
usd_to_inr = 83.0
df_stocks['sp500_inr'] = df_stocks['^GSPC'] * usd_to_inr
df_stocks['nasdaq_inr'] = df_stocks['^IXIC'] * usd_to_inr
df_stocks['nifty50_inr'] = df_stocks['^NSEI']

# 5. Keep only the final columns
final_df_stocks = df_stocks[['date', 'sp500_inr', 'nasdaq_inr', 'nifty50_inr']]

# 6. Save and Download
file_name = 'stock_prices_2020_2025_inr.csv'
final_df_stocks.to_csv(file_name, index=False)
files.download(file_name)

print("---------------------------------------")
print("MISSION COMPLETE: Your Stock data is ready in INR!")
```

```python
import yfinance as yf
import pandas as pd
from google.colab import files

# 1. Setup Parameters
# Tickers: S&P 500, NASDAQ, NIFTY 50
tickers = ["^GSPC", "^IXIC", "^NSEI"]
start_date = "2020-01-01"
end_date = "2025-09-30"
usd_to_inr = 83.0  # Constant conversion rate for consistency

print(f"Fetching data from Yahoo Finance for {tickers}...")

# 2. Extract Data
# We grab the 'Close' price specifically
raw_data = yf.download(tickers, start=start_date, end=end_date)['Close']

# 3. Process and Convert to INR
df_stocks = raw_data.reset_index()

# Rename columns for clarity (Yahoo returns tickers as column names)
# Note: Nifty 50 (^NSEI) is already in INR
df_stocks.columns = ['date', 'sp500_usd', 'nasdaq_usd', 'nifty50_inr']

# Create the converted INR columns for US indices
df_stocks['sp500_inr'] = df_stocks['sp500_usd'] * usd_to_inr
df_stocks['nasdaq_inr'] = df_stocks['nasdaq_usd'] * usd_to_inr

# 4. Clean Data: Drop NaN values
# This removes rows where any market was closed (holidays/weekends)
df_clean = df_stocks[['date', 'sp500_inr', 'nasdaq_inr', 'nifty50_inr']].dropna()

# 5. Save and Download
file_name = 'stock_prices_2020_2025_clean_inr.csv'
df_clean.to_csv(file_name, index=False)
files.download(file_name)

print("---------------------------------------")
print(f"Success! Cleaned data saved with {len(df_clean)} rows.")
print(f"Data range: {start_date} to {end_date}")
```
